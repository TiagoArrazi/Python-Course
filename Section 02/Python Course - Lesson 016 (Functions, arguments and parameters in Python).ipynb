```python
def check_primes(limit):
    for n in range(2,limit):
        check_if_prime(n)

def check_if_prime(number):
    for x in range(2, number):
        if number % x == 0:
            print(f'{number} equals {x} * {number//x}')
            break
    else:
        print(f'{number} is a prime number')
        
    
check_primes(100)
```
